```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../data/bronze/ml2_student_performance.csv')
df.head(5)

plt.figure(figsize=(8,6))
df['Pass/Fail (1=Pass, 0=Fail)'].hist(bins=2, color='skyblue', edgecolor='black', rwidth=0.8)
plt.title('Histogram of Pass/Fail Outcomes')
plt.xlabel('Pass/Fail')
plt.xticks([0, 1], labels=['Fail', 'Pass'])
plt.ylabel('Frequency')
plt.grid(False)
plt.show()
```

```python
#Switching the values in Pass/Fail column since the target is the minority

df['Pass/Fail (1=Pass, 0=Fail)'] = df['Pass/Fail (1=Pass, 0=Fail)'].apply(lambda x: 1 if x == 0 else 0)
```

```python
plt.figure(figsize=(8,6))
df['Pass/Fail (1=Pass, 0=Fail)'].hist(bins=2, color='skyblue', edgecolor='black', rwidth=0.8)
plt.title('Histogram of Pass/Fail Outcomes')
plt.xlabel('Pass/Fail')
plt.xticks([0, 1], labels=['Pass', 'Fail'])
plt.ylabel('Frequency')
plt.grid(False)
plt.show()
```
